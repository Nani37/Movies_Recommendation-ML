```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/refs/heads/main/Movies%20Recommendation.csv')
```

```python
print(df.head())
df.info()
print(df.shape)
print(df.columns)
```

```python
df_features = df[['Movie_Genre', 'Movie_Keywords', 'Movie_Tagline', 'Movie_Cast', 'Movie_Director']]
print(df_features.shape)
df_features = df[['Movie_Genre', 'Movie_Keywords', 'Movie_Tagline', 'Movie_Cast', 'Movie_Director']].fillna('')
x = df_features['Movie_Genre'] + ' ' + df_features['Movie_Keywords'] + ' ' + df_features['Movie_Tagline'] + ' ' + df_features['Movie_Cast'] + ' ' + df_features['Movie_Director']
x = x.astype(str)
tfidf = TfidfVectorizer()
x = tfidf.fit_transform(x)

print(x.shape)
```

```python
similarity_score = cosine_similarity(x)
print(similarity_score.shape)
```

```python
# User input for favorite movie
Favourite_Movie_Name = input('Enter your favourite movie name: ')

# Find the closest match to the user's favorite movie
All_Movies_Name = df['Movie_Title'].values
Movie_Recommendation = difflib.get_close_matches(Favourite_Movie_Name, All_Movies_Name)
print(Movie_Recommendation)
```

```python
# Get the first closest match
if Movie_Recommendation:
    Close_Match = Movie_Recommendation[0]
    print(Close_Match)

    # Find the index of the closest match movie
    Index_of_Close_Match_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
    print(Index_of_Close_Match_Movie)

    # Calculate recommendation scores for all movies based on the closest match
    Recommendation_Score = list(enumerate(similarity_score[Index_of_Close_Match_Movie]))

    # Sort the movies based on similarity scores
    Sorted_Similar_Movies = sorted(Recommendation_Score, key=lambda x: x[1], reverse=True)

    # Print the top 30 movie recommendations
    print('Top 30 Movies Suggested for you: \n')
    i = 1
    for movie in Sorted_Similar_Movies:
        index = movie[0]
        title_from_index = df[df.index == index]['Movie_Title'].values[0]
        if i < 31:
            print(i, '.', title_from_index)
            i += 1
else:
    print("No match found for the movie name.")
```
